```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import train_test_split
import itertools
import joblib
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
data = pd.read_csv('C:/ML_Data/Dataset_Vietnam/Bac Lieu.csv', parse_dates=['date'])
data.set_index('date', inplace=True)
data = data.sort_index()

# Gán tần suất cho dữ liệu
data = data.asfreq('D')  # Đảm bảo chuỗi thời gian liên tục với tần suất hàng ngày

# Kiểm tra và xử lý giá trị thiếu
data = data.fillna(method='pad')  # Điền giá trị thiếu bằng phương pháp 'pad'

# Biến mục tiêu và biến ngoại lai
y = data['humidi']  # Biến dự báo
X = data[['cloud', 'max', 'rain']]  # Biến ngoại lai

# Chia dữ liệu thành tập huấn luyện và kiểm tra
train_size = int(len(data) * 0.8)
train_y, test_y = y[:train_size], y[train_size:]
train_X, test_X = X[:train_size], X[train_size:]

# Phạm vi giá trị p, d, q để thử nghiệm
p = range(0, 4)  # Giá trị AR
d = range(0, 2)  # Sai phân
q = range(0, 4)  # Giá trị MA

# Tạo tổ hợp các giá trị p, d, q
pdq = list(itertools.product(p, d, q))

# Lưu kết quả tốt nhất
best_aic = float("inf")
best_bic = float("inf")
best_params = None
best_model = None

# Lặp qua từng tổ hợp (p, d, q) để tìm tham số tối ưu
for params in pdq:
    try:
        model = ARIMA(train_y, exog=train_X, order=params)
        model_fit = model.fit()
        aic = model_fit.aic
        bic = model_fit.bic

        # Cập nhật mô hình nếu AIC hoặc BIC tốt hơn
        if aic < best_aic:
            best_aic = aic
            best_params = params
            best_model = model_fit
        if bic < best_bic:
            best_bic = bic

        print(f"ARIMAX{params} - AIC: {aic:.2f}, BIC: {bic:.2f}")
    except Exception as e:
        print(f"Model ARIMAX{params} không khả thi: {e}")

# In kết quả tốt nhất
print(f"\nBest Model: ARIMAX{best_params} with AIC: {best_aic:.2f}, BIC: {best_bic:.2f}")

# Dự báo trên tập kiểm tra
forecast = best_model.forecast(steps=len(test_y), exog=test_X)

# Đánh giá mô hình
mae = mean_absolute_error(test_y, forecast)
rmse = mean_squared_error(test_y, forecast, squared=False)
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

# Vẽ biểu đồ dự báo
plt.figure(figsize=(10, 6))
plt.plot(test_y.index, test_y, label='Giá trị thực tế', color='blue')
plt.plot(test_y.index, forecast, label='Dự báo ARIMAX', color='red', linestyle='--')
plt.xticks(rotation=45)
plt.title('Dự báo độ ẩm với ARIMAX')
plt.xlabel('Ngày')
plt.ylabel('Độ ẩm (Humidity)')
plt.legend()
plt.tight_layout()
plt.show()

# Lưu mô hình tốt nhất và các thông tin liên quan
model_data = {
    "model": best_model,
    "best_params": best_params,
    "best_aic": best_aic,
    "best_bic": best_bic
}
model_path = 'C:\ML_Data\Model_ARIMAX\Bac_Lieu_arimax.pkl'
joblib.dump(model_data, model_path)
print(f"Mô hình tốt nhất đã được lưu tại: {model_path}")
```

## Dự đoán cho ngày kế tiếp

```python
import pandas as pd
import joblib

# Tải mô hình đã lưu
model_data = joblib.load('C:/ML_Data/Model_Vietnam/Bac_Lieu_arimax.pkl')
best_model = model_data["model"]

# Dữ liệu các yếu tố ngoại lai cho ngày tiếp theo
next_day_exog = pd.DataFrame({
    "cloud": [50],   # Độ che phủ mây (giá trị ví dụ)
    "max": [32],     # Nhiệt độ cao nhất (giá trị ví dụ)
    "rain": [0.2]    # Lượng mưa (giá trị ví dụ)
})

# Dự đoán độ ẩm cho ngày tiếp theo
forecast_next_day = best_model.forecast(steps=1, exog=next_day_exog)
print(f"Dự báo độ ẩm cho ngày tiếp theo: {forecast_next_day[0]:.2f}")
```
